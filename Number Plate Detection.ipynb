## Introduction

This project aims to detect number plates in images using the YOLOv8 object detection model. YOLO (You Only Look Once) is a state-of-the-art, real-time object detection system. YOLOv8 is the latest version, offering improved accuracy and speed.

## Dataset

The dataset consists of approximately 1,000+ images, each labeled with the location of number plates. The dataset is split into three parts: training, validation, and testing.

- **Training Set:** 85% of the dataset
- **Validation Set:** 9% of the dataset
- **Testing Set:** 6% of the dataset

## Using Paddle OCR

```python
# Install Paddle OCR

!pip install paddlepaddle
!pip install paddleocr 
```

```python
# Initialize PaddleOCR
ocr = PaddleOCR(use_angle_cls=True, lang='en')  # Specify the language you want to use
```

```python
# Set the folder path containing the images
folder_path = 'car_testing' # give a folder of different images
```

```python
# Get the list of all image files in the folder

image_files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f)) and f.lower().endswith(('.jpg', '.png', '.jpeg', '.webp'))]
```

### Displaying 5 random images from folder

```python
# Number of random images to display
num_random_images = 5

# Select a random sample of image files
random_images = random.sample(image_files, min(num_random_images, len(image_files)))

# Create a figure with subplots
num_rows = (len(random_images) + 1) // 2  # Calculate number of rows needed
fig, axes = plt.subplots(nrows=num_rows, ncols=2, figsize=(12, 12))

# Display each random image
for i, image_file in enumerate(random_images):
    image_path = os.path.join(folder_path, image_file)
    image = cv2.imread(image_path)
    
    if image is None:
        continue

    # Convert BGR image to RGB for display
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Make predictions using the YOLO model
    results = model(image_path)
    
    # Draw bounding boxes on the image
    for result in results:
        if len(result.boxes) > 0:
            for box in result.boxes:
                x1, y1, x2, y2 = map(int, box.xyxy[0])
                cv2.rectangle(image_rgb, (x1, y1), (x2, y2), (0, 255, 0), 2)
    
    # Determine the subplot location
    row = i // 2
    col = i % 2
    axes[row, col].imshow(image_rgb)
    axes[row, col].axis('off')
    axes[row, col].set_title(image_file)

# Hide any unused subplots
for j in range(len(random_images), num_rows * 2):
    axes[j // 2, j % 2].axis('off')

plt.tight_layout()
plt.show()
```

## Extracting text from number plat

```python
# List to store results
results_list = []

for image_file in image_files:
    image_path = os.path.join(folder_path, image_file)
    
    # Load the image
    image = cv2.imread(image_path)
    if image is None:
        print(f"Error: Could not load image {image_file}")
        continue

    # Make predictions using the YOLO model
    results = model(image_path)
    
    text_lines = []
    number_plate_detected = False

    # Iterate over the results
    for result in results:
        if len(result.boxes) == 0:
            print(f"No number plates detected in {image_file}")
            continue
        
        number_plate_detected = True

        # Extract text from each bounding box
        for box in result.boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            cropped_image = image[y1:y2, x1:x2]
            
            # Convert cropped image to RGB for PaddleOCR
            cropped_image_rgb = cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB)

            # Apply OCR to the cropped image
            ocr_results = ocr.ocr(cropped_image_rgb, cls=True)
            
            # Collect the detected text
            for line in ocr_results[0]:
                _, (text, confidence) = line
                text_lines.append(text)  # Collect all detected texts

    # Join all detected lines into a single string
    full_text = ' '.join(text_lines) if text_lines else 'No text detected'

    # Add the results to the list
    results_list.append((image_file, full_text))

# clear output
clear_output(wait=True)

# Create a DataFrame to store the extracted license plate texts
results_df = pd.DataFrame(results_list, columns=['Filename', 'Extracted Text'])

# Display the DataFrame
print(results_df)
```

```python
# Saving above resuluts_df

results_df.to_csv("extracted_text", index=False) # "indext=False" to save CSV file without including row indices
```
